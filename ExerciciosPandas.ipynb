1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd

dados = pd.read_csv('dados/alunos3.csv', sep=";")

dados["Prova_1"] = dados["Prova_1"].str.replace(',','.').astype(float)
dados["Prova_2"] = dados["Prova_2"].str.replace(',','.').astype(float)
dados["Prova_3"] = dados["Prova_3"].str.replace(',','.').astype(float)
dados["Prova_4"] = dados["Prova_4"].astype(float)

dados["Media_provas"] = (dados["Prova_1"] + dados["Prova_2"] + dados["Prova_3"] + dados["Prova_4"])/4
dados["Media_provas"] = dados["Media_provas"].round(2)
dados
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_nota = dados["Media_provas"].max()
selecao_1 = (dados["Media_provas"] == maior_nota)
dt_maior_nota = dados[selecao_1]
print('A maior nota é de: ' + dt_maior_nota["Nome"])

menor_nota = dados["Media_provas"].min()
selecao_2 = (dados["Media_provas"] == menor_nota)
dt_menor_nota = dados[selecao_2]
print('A Menor nota é de: ' + dt_menor_nota["Nome"])
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
import pandas as pd

dados_2 = pd.read_excel('dados/cadastro_alunos.xlsx')

df = pd.merge(dados, dados_2, how = 'inner', on = 'RA')
df
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_1 = df.groupby('Sexo')['Media_provas'].mean()
df_1 = pd.DataFrame(df_1)
df_1
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
selecao = (df['Media_provas'] >= 7)

df_2 = df[selecao]
df_2 = pd.DataFrame(df_2)['Idade'].mean()
print("A média de idade dos alunos aprovados é " + str(df_2) + " anos")
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df_3 = df.groupby('Cidade')['Media_provas'].mean().round(2)
df_3 = pd.DataFrame(df_3)

maior_nota_6 = df_3["Media_provas"].max().round(2)

selecao_3 = (df_3['Media_provas'] >= maior_nota_6)
n4 = df_3[selecao_3].reset_index()

print("Maior média é " + n4["Cidade"] + " com " + str(maior_nota_6))
```

7. Leia o arquivo `aluno3.csv`, apenas as colunas RA, Prova_1, Prova_2, Prova_3, Prova_4 calcule a médias das notas da prova e salve em um arquivo excel (xlsx) as colunas RA e Média.


```python
import pandas as pd

dados = pd.read_csv('dados/alunos3.csv', sep=";")

dados["Prova_1"] = dados["Prova_1"].str.replace(',','.').astype(float)
dados["Prova_2"] = dados["Prova_2"].str.replace(',','.').astype(float)
dados["Prova_3"] = dados["Prova_3"].str.replace(',','.').astype(float)
dados["Prova_4"] = dados["Prova_4"].astype(float)

dados["Media"] = (dados["Prova_1"] + dados["Prova_2"] + dados["Prova_3"] + dados["Prova_4"])/4
dados["Media"] = dados["Media"].round(2)
dados = dados[['RA', 'Media']]
dados

dados.to_excel('dados/novo_cadastro_alunos.xlsx', index = False)
```
